2. Descriptive Statistics: Calculate basic statistics (mean, median, mode, standard deviation).

```python
print("Retail Sales Summary Statistics")
d1=df_retail.describe()
print(d1,"\n")

Median_Age=df_retail['Age'].median()
print("Median Age=",Median_Age)
```

```python
print("McDonald Menu Summary Statistics")
df_menu.describe().T
```

```python

x=df_menu.select_dtypes(include=np.number).columns

stat_menu={
    col: {
        'Means': df_menu[col].mean(),
        'Median': df_menu[col].median(),
        'Mode':df_menu[col].mode().iloc[0],
        'Std Dev':df_menu[col].std()
    }
    for col in x 
}
y=pd.DataFrame(stat_menu).T
y.round(2)
```

3.Time Series Analysis: Analyze sales trends over time using time series techniques.

```python
df_retail.columns
```

```python
df_menu.columns
```

```python
df_retail['Date']
```

```python
df_retail['Date']=pd.to_datetime(df_retail['Date'])
```

```python
df_retail.set_index('Date',inplace=True)
df_retail.sort_index(inplace=True)
```

```python
df_retail.head()
```

```python
## Plotting Monthly Sales Trend

monthly_sales = df_retail['Total Amount'].resample('ME').sum()

plt.figure(figsize=(12,6))
plt.plot(monthly_sales, marker='o', linestyle='-', color='blue')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales Amount')
plt.grid(True)
plt.show()

```

```python
monthly_Quantity = df_retail['Quantity'].resample('ME').sum()

plt.figure(figsize=(12,6))
plt.plot(monthly_Quantity, marker='o', linestyle='--', color='red')
plt.title('Monthly Quantity Trend')
plt.xlabel('Month')
plt.ylabel('Total Qunatity')
plt.grid(True)
plt.show()
```

4. Customer and Product Analysis: Analyze customer demographics and purchasing behavior.

```python
# Count of Male vs Female
sns.countplot(data=df_retail, x='Gender', hue='Gender', palette='Set2')
plt.title("Customer Gender Distribution")
plt.ylabel("Number of Customers")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=df_retail, x='Age', kde=True, bins=20,  color='red')
plt.title("Age Distribution of Customers")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# Most Popular Product

plt.figure(figsize=(10,6))
sns.countplot(data=df_retail, x='Product Category', hue='Product Category', palette='cool', legend=False)
plt.title("Top product Categories (by Quantity Sold)")
plt.xlabel("Count")
plt.ylabel("Product Category")
plt.show()
```

```python
# Total Sales By Product Category
plt.figure(figsize=(10,6))
sns.barplot(y='Total Amount', x='Product Category', hue='Product Category', data=df_retail, palette="Set3")
plt.title("Total Sales by product category")
plt.ylabel("Total Amount($)")
plt.xlabel("Product Category")
plt.show()
```

```python
### Menu.csv
# Counts of Items by Category

plt.figure(figsize=(15,6))
sns.countplot(data=df_menu,y='Category', hue='Category',palette='Set2')
plt.title("Number of Items per Category")
plt.xlabel=("Count")
plt.ylabel("Category")
plt.show()
```

```python
# Average Calories per Food Category
avg_cal = df_menu.groupby('Category')['Calories'].mean()
plt.figure(figsize=(15,6))
sns.barplot(y=avg_cal.index, x=avg_cal.values,hue=avg_cal.values, palette="Set3" , legend=False)
plt.title("Average Calories per Food Category")

plt.show()
```

```python
#Top 10 Highest calorie Items
top_10 = df_menu[['Item', 'Calories']].sort_values(by='Calories', ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(data=top_10, y='Item', x='Calories', palette='magma')
plt.title("Top 10 Highest Calorie Items")
plt.show()
```

5.Visualization: Present insights through bar charts, line plots, and heatmaps.

```python
## Retail Dataset
# Calculate correlation matrix
corr=  df_retail[['Age','Quantity','Price per Unit','Total Amount']].corr()
plt.figure(figsize=(16,6))
sns.heatmap(data=corr,annot=True, cmap='coolwarm',fmt=".2f")
plt.title("Correlation Heatmap (Retail Dataset)")
plt.show()
```

```python
# Sales By product Category by Pie Chart

Sales = df_retail.groupby('Product Category')['Total Amount'].sum()
plt.figure(figsize=(16,6))
plt.pie(Sales, labels=Sales.index, colors=sns.color_palette('magma'), autopct='%1.1f%%')
plt.title("Sales Share by product Category")
plt.show()
```

```python
# Gender Vs Average Spend

Avg_Spn=df_retail.groupby('Gender')['Total Amount'].mean()
print(Avg_Spn)
plt.figure(figsize=(6,5))
Avg_Spn.plot(kind='bar',color=['hotpink','royalblue'])
plt.title("Gender Vs Average Spend")
plt.show()
```

```python
## Menu.csv
# Correlation Heatmap
cols=['Calories','Saturated Fat','Cholesterol','Sodium','Carbohydrates','Dietary Fiber','Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)',
       'Calcium (% Daily Value)', 'Iron (% Daily Value)']
plt.figure(figsize=(16,6))
sns.heatmap(df_menu[cols].corr(),annot=True,cmap='coolwarm')
plt.title("Nutrient Correlation HeatMap")
plt.show()
```

```python
df_menu.columns
```

```python
# Item Distribution By Category by PieChart
Itm_dis=df_menu.groupby('Category')['Item'].count()
plt.figure(figsize=(10,6))
Itm_dis.plot(kind='pie', colors=sns.color_palette('Set2'),autopct='%1.1f%%')
plt.title("Item Distribution By Category")
plt.show()

```

```python
## Category Vs Avg Protein by Bar Chart

avg_protein = df_menu.groupby('Category')['Protein'].mean().sort_values()
plt.figure(figsize=(16,6))
avg_protein.plot(kind='barh', color=sns.color_palette('magma'))
plt.title("Category Vs Avg Protein")
plt.show()
```
